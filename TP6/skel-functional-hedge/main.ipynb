```python
### Q6 : 

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from scipy.stats import norm 
data = pd.read_csv("pnl.txt" ,header=None)
data.head()

data[0].plot.hist(density=True)
mu = data[0].mean()
sigma = data[0].std()
x = np.linspace(data[0].min() , data[0].max() , 1000)
plt.plot(x , norm.pdf(x , mu , sigma) , label="loi normale" )
plt.legend()
plt.show()
```
